#### **B. From the Integration of Princely States to the Dissolution of the British Commonwealth** 

The modern conception of British forces in India arose during a critical juncture marked by the British Empire's intense rivalry with France in the mid-1750s. Lord Wellesley, a prominent British statesman and colonial administrator, spearheaded a pivotal phase of British expansion in India. Recognizing the strategic imperative of consolidating British influence, Lord Wellesley, along with subsequent Governor Generals, embarked on an ambitious campaign to integrate numerous princely states into the British colonial fold. This concerted effort aimed to bolster British control over the Indian subcontinent, securing vital territories and resources amidst the backdrop of geopolitical competition. Under Wellesley's leadership and the subsequent administrations, diplomatic maneuvers, military conquests, and political negotiations were employed to bring princely states under British suzerainty. This expansionist drive not only fortified British dominance in India but also reshaped the political landscape, paving the way for the consolidation of British colonial rule over vast swathes of the subcontinent. Lord Wellesley's strategic vision and the subsequent efforts of his successors underscored the pivotal role of princely state integration in shaping the trajectory of British imperialism in India during this era of geopolitical tumult. Below is a brief description of the four phases of Wellesley's integrative mission. 

**I. The Ring Fence**

The first policy, the Ring Fence, was implemented between 1765 and 1813, marking the beginning of Britain's annexation strategy. This policy was crafted as a response to the perceived threat posed by the formidable coalition of the Marathas, Hyderabad, and Mysore, aiming to safeguard the British East India Company's frontiers (Verghese, 2019). To achieve this objective, buffer zones were established utilizing neighboring states, effectively creating a defensive perimeter around Company territories. Notably, the policy also aimed to protect British frontiers in the northwest from potential Afghan incursions. In regions like Awadh State, where the danger from the Marathas and Afghans was pronounced, the Ring Fence policy was rigorously enforced, with the Nawab mandated to finance the defense efforts as per the Company's stipulations. The defense of Awadh held particular significance for safeguarding Bengal, emphasizing the strategic importance of these buffer states in the broader colonial defense strategy. As a result, the princely states were effectively encircled within the Ring Fence, obliged to bear the financial burden of their own security against external threats, albeit with the military support of the Company — a policy that laid the groundwork for the subsequent adoption of the subsidiary alliance system. Although the East India Company abided by a noninterventionist policy, the private corporation continued to wage war against Indian rulers with smaller military presence, winning significant territory across the continent and forming alliances with smaller tribal communities. Due to these smaller conquests, the British grew in prominence as the subcontinent’s political power, leading to the second phase of British domination.

**II. The Subsidiary Alliance**

The second phase, the Subsidiary Alliance, introduced in 1798 during Lord Wellesley's tenure as Governor-General of India, marked a significant offensive shift in British colonial policy on the Indian subcontinent. Prior to 1760, the British East India Company had primarily engaged in treaties with local coastal powers for commercial purposes and piracy suppression. Under the Subsidiary Alliance system, the Company negotiated agreements with Indian princely states, beginning with the Nizam of Hyderabad in 1798, followed by treaties with Oudh, Cooch Behar, and Carnatic. The terms of these alliances were structured to consolidate British control and influence over the princely states while ostensibly preserving their sovereignty. Key provisions included the deployment of well-trained British troops within the princely state, with the state bearing the cost of their maintenance through an annual subsidy. Additionally, princely states were prohibited from forming alliances or treaties with other Indian kingdoms without prior consent from the Company. Furthermore, a British resident was stationed within the princely state to oversee affairs and ensure compliance with the terms of the alliance. Under the Subsidiary Alliance system, princely states were also restricted in their ability to employ Europeans and were required to obtain Company approval for such appointments. However, the alliances purportedly respected the internal sovereignty of the princely states, with the Company agreeing not to interfere in their internal affairs. Despite this assurance, the presence of British troops and residents effectively curtailed the autonomy of the princely states, placing them under de facto British control. The Subsidiary Alliance system represented a strategic tool for expanding British influence and military dominance on the Indian subcontinent, ultimately laying the groundwork for the further consolidation of British colonial rule in India.

**III. Subordinate Isolation**

Under the third phase of colonization, the policy of Subordinate Isolation, princely states were expected to cooperate in a subordinate manner with the British government and acknowledge its supremacy. While the princely states retained sovereignty in internal administration, they surrendered all forms of external sovereignty, effectively becoming subordinate entities within the British colonial framework. Central to the implementation of the Subordinate Isolation policy was the transformation of British residents from mere diplomatic agents of foreign powers to executive and controlling officers representing the superior British government. This shift in the role of British residents underscored the consolidation of British control over princely states and the extension of British administrative authority, introducing the landlord-revenue system and marking the precursory extractionist environment in which direct and indirect districts emerged.

The enactment of the Charter Act of 1833 marked a significant milestone in the evolution of British colonial governance in India. While the act abolished the British East India Company's commercial functions, it retained its political functions, solidifying the Company's role as the paramount political authority in India. In the ensuing years, the Company increasingly asserted its authority over princely states, particularly in matters of succession. The Company began insisting on prior approval or sanction for all succession-related matters, exerting greater control over the internal affairs of princely states. From 1834 onwards, the Company's Board of Directors adopted a more aggressive stance towards annexation, pushing for the incorporation of princely states into British territories whenever and wherever possible. This policy of annexation reached its zenith through the Doctrine of Lapse, initiated by Governor General Lord Dalhousie between 1848 to 1856, which allowed the British to annex princely states if they lacked a male heir or failed to meet certain conditions. Through this doctrine, the kingdoms of Satara (1848), Sambalpur (1850), Udaipur (1852), Nagpur (1853), and Jhansi (1854) all collapsed under British control. Overall, the policy of Subordinate Isolation represented a pivotal phase in British colonial administration in India, characterized by the consolidation of British control and the gradual erosion of princely autonomy, ultimately leading to the incorporation of princely states into the British Empire.

**IV. The Sepoy Mutiny**

The period of annexation was largely challenged by the Sepoy Mutiny, also known as the Indian Rebellion of 1857, which had a significant impact on British colonial policies in India, including the policy of annexation. While the immediate outcome of the rebellion did not directly end the policy of annexation, it did lead to a reevaluation of British colonial strategies and a shift in approach towards governing India. The rebellion erupted primarily due to a confluence of factors, including grievances among Indian soldiers (sepoys) in the British East India Company's army, discontent among various sections of Indian society over British policies and social reforms, and underlying tensions related to cultural, religious, and economic issues. The uprising quickly spread across northern and central India, posing a serious challenge to British authority. In response to the rebellion, the British government took decisive military action to suppress the uprising, deploying additional troops and imposing martial law in affected regions. The rebellion was eventually quelled by British forces, but not without significant bloodshed and loss of life on both sides.


While the Sepoy Mutiny did not directly result in the cessation of the policy of annexation, it did prompt the British government to reconsider its approach to governing India. The rebellion exposed the vulnerabilities of British rule and highlighted the need for reforms in administration, military organization, and social policies. Following the suppression of the rebellion, the British government dissolved the British East India Company and transferred control of India directly to the British Crown, establishing the British Raj. This transition marked a shift towards more direct and centralized control over Indian affairs, as well as a reassessment of colonial policies. In the aftermath of the rebellion, the British government adopted a more cautious and pragmatic approach towards expansion and annexation in India. While annexations did continue in some instances, particularly in regions where British interests were threatened or strategic considerations warranted intervention, there was greater emphasis on consolidation of existing territories and maintaining stability rather than aggressive expansion. Overall, while the Sepoy Mutiny did not directly result in the cessation of the policy of annexation, it did contribute to a broader reevaluation of British colonial policies and strategies in India, leading to changes in governance and administration that ultimately shaped the trajectory of British rule in the subcontinent.

**V. The Rise of the Independence Movement**

Over the subsequent decades, Indian nationalist movements began to emerge, advocating for greater political representation, civil liberties, and ultimately, independence from British rule. Leaders such as Dadabhai Naoroji, Bal Gangadhar Tilak, Lala Lajpat Rai, and Mohandas Gandhi played pivotal roles in galvanizing Indian public opinion and mobilizing support for the nationalist cause. The First World War (1914-1918) and the Second World War (1939-1945) also had profound effects on British India. Indian troops played a significant role in both conflicts, and their contributions highlighted the contradictions of colonial rule, where Indians were asked to fight and die for a British empire that denied them full political rights and self-determination. The post-World War II period saw a surge in nationalist sentiment in India, fueled by widespread dissatisfaction with British rule and growing demands for independence. The British government, under pressure from both Indian nationalists and international opinion, recognized the need to address India's aspirations for self-rule.

Several factors contributed to the British decision to leave India, including the rise of nationalist movements, international pressure, economic considerations, and the changing geopolitical landscape following the world wars. Ultimately, the quest for Indian independence and self-determination proved to be insurmountable challenges for British colonialism in the subcontinent, leading to the eventual withdrawal of British forces and the establishment of the Indian Independence Act in 1947, which paved the way for the partition of British India into two separate independent nations: India and Pakistan. On August 15, 1947, India gained independence, marking the end of nearly two centuries of British colonial rule.

#### **C. How the Landlord System During Subordinate Isolation Catalyzed the Direct-Indirect Divide**

The discussion of direct and indirect British control is heavily influenced by the establishment of the landlord-based revenue system during the Subordinate Isolation phase of British colonization. Pioneering early investigations into the impact of British occupation in India, Banerjee and Iyer (2005) focused their research on the landlord revenue systems, a system of land taxation and administration established by the British colonial government among coastal states and districts that lacked a direct British presence. The most prominent revenue system implemented by the British was the Zamindari System, also known as the Permanent Settlement.

Under the Zamindari Settlement, introduced in Bengal, Bihar, and Orissa by the Cornwallis Code of 1793, the British government recognized certain landlords, known as zamindars, as the hereditary owners of land who were loyal to the British extraction agenda in India. These zamindars were typically white members of the local aristocracy or landed gentry, who served as intermediaries between the British authorities and the local agricultural communities. The zamindars were given permanent hereditary rights to collect land revenue from the peasants who cultivated the land. In return, the zamindars were required to pay a fixed amount of revenue to the British government, irrespective of fluctuations in agricultural output or changes in land productivity.

This system aimed to provide stability and predictability in revenue collection for the colonial administration, while also incentivizing the zamindars to invest in land improvements and agricultural development. However, in practice, the Permanent Settlement often led to exploitation and impoverishment of the peasants, as the zamindars had considerable power over their tenants and frequently extracted exorbitant rents. In fact, Iyer and Banerjee (2010) documented the negative effects of the landlord-based system on agricultural investments and public good provision, theorizing its contribution to the establishment of an entrenched elite that fostered a class-based divide and societal conflicts that hindered collective action for public goods provision. Their research is corroborated by Pandey (2010) who compared socio-economic outcomes between landlord and non-landlord districts in Uttar Pradesh. Pandey confirmed that villages in non-landlord districts showed greater performance on governance and educational metrics.


Apart from the Permanent Settlement, other revenue systems were also implemented in different regions of India, including the Ryotwari System and the Mahalwari System. These systems varied in their administration and mechanisms of revenue collection but shared the overarching goal of maximizing revenue extraction for the British colonial government while relying on intermediaries, such as landlords or individual cultivators, for revenue collection at the local level.

Over time, districts that faced revenue-driven administrations became proxies for states ruled under strict British influence, known as "directly" ruled districts in this paper. As landlord-based revenue systems largely appeared in directly-ruled districts and non-landlord-based systems aligned with nativist states, the landlord system remains essential to account for in the regression analysis. Recognizing this correlation becomes crucial for a comprehensive interpretation of the results, as it provides context for the motivation behind the establishment of directly ruled territories.

#### **D. The Legacies of Direct and Indirect Colonial Rule**

British colonialism is a largely studied topic. The question of why areas directly governed by a colonial power exhibit different long-term outcomes compared to those ruled indirectly has been less studied. Two overarching rationales have historically been considered when analyzing the impact of direct and indirect colonialism on longterm development: How historical circumstances residual from colonial pasts shape A) intra-population interactions and B) institutional hypotheses that influence future development in the region.

Considering the first case, how historical circumstances have shaped populations through public policy, researchers find that historical circumstances often lead to bias in postcolonial policymaking which severely infiltrates legislative and judicial spheres and results in inequitable policy decisions. Examples from cross-country studies illustrate this, such as the variations in the functioning of legal systems for different social groups across different countries such as in the Middle East and Africa (Rodrik, Subramanian, and Trebbi, 2004). When applied to the Indian context, Iyer explains that the the discernible impact of both direct and indirect rule on democratic processes comes sharply into focus, notably through the persistent involvement of former ruling families of native states in postcolonial Indian politics. The presence of these individuals in the political landscape is theorized to exert influence on voter participation, electorate information, and, consequently, the responsiveness of political representatives to public needs. This underscores the imperative need for analyzing the pre-contextual conditions that shape the intricate interplay between historical legacies and contemporary democratic dynamics in India.

The second question answered from analyzing colonial administrations is the effect of colonial policies on shaping institutional choices and longterm development. Three main frameworks are considered. Firstly, academia surrounding the topic of colonization highlights the intentions behind colonizing missions, across the British, Dutch, French, Spanish, and Portuguese. Across differing colonizing regimes, multiple versions of the extractive hypothesis emerge, suggesting that colonial rulers  resort to excessive taxation or extreme agricultural development in directly ruled areas to extract resources for the colonial power (Besley and Burgess, 2000). Secondly, the institutional hypothesis posits that colonial rulers establish suboptimal institutions in areas where long-term settlement is not intended, suggesting that areas under indirect rule - that is, areas administrated by princely or nativist rulers - tend to pursue more favorable institutional choices designed to increase human capital (Acemoglu et al. 2001). Lastly, Iyer (2010) considers the accountability hypothesis in which native district rulers face significantly more incentive than British administrators to provide quality public goods to district citizens driven by their longterm tenures, which allowed them to invest in longer horizon development projects, and by the fear of deposition from a ruling position. These factors are expected to influence long-term investments, such as schools or health facilities; however, a large-scale study into the effectiveness of this incentive is yet to be conducted.

In conclusion, the investigation into the legacies of direct colonial rule reveals a nuanced landscape deserving in-depth exploration. The underexamined realm of British colonialism, particularly the comparative analysis of areas directly governed and those ruled indirectly, unveils complex dynamics influencing postcolonial outcomes. Researchers highlight the intricate interplay between historical circumstances and postcolonial policies, evident in the Indian context where democratic processes are shaped by the enduring impact of pre-contextual conditions. The ongoing potential for further research in this uncharted territory enriches our understanding of historical dynamics and provides valuable insights for contemporary policy considerations in postcolonial societies.

### **B. Empirical Model**

To test the impact of the British administration style on socio-economic indicators, three empirical models will be tested, with each subsequent regression increasing in precisition. 

#### **Regression 1:**

$ Y_i = \beta_0 + \beta_1 D_i + \theta_1 S_i + U_i $

The first regression model presented in this analysis examines the relationship between the socioeconomic variable of interest $Y_i$ and several explanatory variables, including two binary dummy variables: Dummy variable $D_i$, where $D = 1$ if the district is directly-ruled and $D = 0$ if the district is indirectly-ruled, and dummy variable $S_i$, where $S = 1$ if the district is in Maharastra and $S = 0$ if the district is in Orissa. In the model, the coefficient $ B_1 $ quantifies the impact of the administrative style on the socioeconomic variable. The error term $U_i$ accounts for unobserved factors influencing $Y_i$ not explicitly included in the model. Using this OLS regression framework, a comprehensive assessment of the impact of both binary independent variables on the dependent variable can be seen, providing valuable insights into the socioeconomic relationships under investigation.

#### **Regression 2:**

$ Y_i = \beta_0 + \beta_1 D_i + \theta_1 S_i + \theta_2 X_i + U_i $

The second regression model extends upon the first, keeping the same binary variables $D_i$ and $S_i$. In addition, this regression provides additional fortification to the regression through the inclusion of environmental condition control variables, listed as $X_i$. These control variables, namely the area, temperature, rain, and altitude metrics of districts, serve a crucial role in isolating the relationship between the independent variables of interest and the dependent variable by holding constant the effects of other potentially influential factors. For instance, a confounding variable exists in Regression 1 as areas with ambient weather, abundant rainfall, and lower altitudes tend to be favored by British colonizers for their agricultural potential. This fact can skew the relation between a district's administrative status and its socioeconomic development. By including this control variable, I aim to minimize omitted variable bias and improve the accuracy and reliability of the estimates, helping control for alternative explanations or competing hypotheses and enhancing the internal validity of the analysis. 

#### **Regression 3:**

$ Y_i = \beta_0 + \beta_1 D_i + \beta_2 A_i + \theta_1 S_i + \theta_2 X_i + U_i $

In addition to the variables used in Regressions 1 and Regressions 2, the third regression model provides an additional element of consideration, the potential for spillover across districts. Spillovers can occur in the form of informational, experiential, and legislative practices in directly-ruled states spilling over into nativist states, impacting those district's levels of economic attainment. The inclusion of an adjacency dummy variable, denoted as $A_i$, allows us to test for the potential spillover. Describing the variable, $A = 1$ if an individual district is bordered by a directly-administered district and $A = 0$ if an individual district does not border a directly-administered district lacking such adjacency. $B_2$ then measures the causal effect of adjacency on specific variables of interest.

```python
#all together
#interaction term means that X1s impact on Y actually depends on another variable (D1 for subgroups, or X1 for other factors in general)
#the reason why the beta on the regressions with interaction terms are higher is because the interaction term gives you a JOINT effect of size B3 on the dependent variable. 

import statsmodels.formula.api as smf
from statsmodels.iolib.summary2 import summary_col

# Regression 1
result_1 = smf.ols('testscr ~ str + el_pct + meal_pct', data=df).fit(cov_type='HC1')  # use heteroskedastic robust standard error
#regress test score with linear terms 
#only uses regular multivariable, so each coefficient is = to the PE of that variable holding the others constant.

result_2 = smf.ols('testscr ~ str + el_pct + meal_pct + log_income', data = df).fit(cov_type="HC1")
#gradually add some nonlinearity (log_income)
#there is a change in the STR here so it could indicate that the previous regression had OVB

result_3 = smf.ols('testscr ~ str +  meal_pct + log_income + HiEL*str', data = df).fit(cov_type="HC1")
#added interation term between hiEL and STR - can be re-rewritten from its interaction term form
#can see causal effect on STR as (B1+B3HiEL)STR; so when HiEL=0 the CE is B1 (0.53), but when HiEL channel is turned on the CE will be (B1+B3) = (-0.53-0.577)=-1.1077

#Results are extremely different across 4 and 5 because they include higher order polynomials - but, as we know with polynomials - that doesn't really impact the causal effect because the causal effect is a linear combination!

result_4 = smf.ols('testscr ~ str +  str_square + str_cube + HiEL + meal_pct + log_income', data = df).fit(cov_type="HC1")
#coeff on squared term is still linear tho so we can use OLS
#the causal effect of the polynomials is the derivative of the STR regressors (and you can input a level of X in for which it changes from)
#to find the causal effect (which is actually a linear combination not a B), you can also just plug in 11 and 10 and find the diff

result_5 = smf.ols('testscr ~ str +  str_square + str_cube + HiEL + HiEL:str + HiEL:str_square + HiEL:str_cube + meal_pct + log_income', data = df).fit(cov_type="HC1")
#most complicated one; have linear and interaction terms 

regressor_display_order = ["str", "str_square", "str_cube", "el_pct", "HiEL", "HiEL:str", "meal_pct", "avginc", "log_income", "Intercept"]
#to find the causal effect, you can change all the polynomials to their derivative and then plug in values for STR and whether the Hi-EL channel is turned on

dfoutput = summary_col([result_1, result_2, result_3, result_4, result_5], stars=False, regressor_order=regressor_display_order)
#look at how the results differ
print(dfoutput)
```

```python
# Independent variables
independent_vars = sm.add_constant(dr1951[['british', 'state']])

# Dependent variables
dependent_vars = ['roadpercapita', 'railpercapita', 'sharehigh','shareprime','litrate','lightsperannum','lfpr','imr']

# Results storage
results_table = []

# Loop through each dependent variable
for var in dependent_vars:
    # Prepare data for current dependent variable
    dependent_var = dr1951[var]
    
    # Fit OLS regression model
    model = sm.OLS(dependent_var, independent_vars)
    result = model.fit()

    # Store results summary
    results_table.append({
        'Independent Variable': 'British Coef.',
        'Dependent Variable': var,
        'Coefficient': result.params['british'],
        'Std. Error': result.bse['british'],
        't-value': result.tvalues['british'],
        'p-value': result.pvalues['british']
    })

    results_table.append({
        'Independent Variable': 'State Control Coef.',
        'Dependent Variable': var,
        'Coefficient': result.params['state'],
        'Std. Error': result.bse['state'],
        't-value': result.tvalues['state'],
        'p-value': result.pvalues['state']
    })

# Convert results to DataFrame
results_table_dr1951 = pd.DataFrame(results_table)

results_table_dr1951.set_index('Dependent Variable', inplace=True)

print(results_table_dr1951)

# Plot coefficients
plt.figure(figsize=(10, 6))
for var in dependent_vars:
    plt.barh(var + ' - British', results_table_dr1951.loc[var, 'Coefficient'], xerr=results_table_dr1951.loc[var, 'Std. Error'], color='skyblue', label='British')
    plt.barh(var + ' - State Control', results_table_dr1951.loc[var, 'Coefficient'], xerr=results_table_dr1951.loc[var, 'Std. Error'], color='salmon', label='State Control')

plt.xlabel('Coefficient')
plt.ylabel('Dependent Variable')
plt.title('Regression Coefficients')
plt.legend()
plt.tight_layout()
plt.show()
```
